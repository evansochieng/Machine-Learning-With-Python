```python
#import libraries
import numpy as np
import pandas as pd
from sklearn.svm import SVC
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read the data
data = pd.read_csv("C:/Users/HP/Desktop/DATA SCIENCE/Python/Learn Python With Rune/Machine Learning With Python/MachineLearningWithPython-main/files/data_02.csv")
data.head()
```

```python
#Make class IDs
class_ids = {"Dobermann":0, "German Shepherd":1, "Rottweiler":2}

#create a column for the ids
data["Class ID"] = data["Class"].apply(lambda x: class_ids[x])
data.head()
```

```python
#Scatter plot the data
fig, ax = plt.subplots()

ax.scatter(x=data["Weight"], y=data["Height"], c=data["Class ID"])
```

```python
#Fit a model
x = data[["Weight", "Height"]]
y = data["Class ID"]

clf = SVC(kernel="linear")
clf.fit(x, y)
```

```python
#Map out the classification

#Create a random selection of the data and shift it to match the axis
x_test = np.random.rand(10000, 2)
x_test = x_test*[40, 20] + [25, 55]

#Predict the random selection of data
y_pred = clf.predict(x_test)

#craete a plot with the data and predictions
fig, ax = plt.subplots()

ax.scatter(x_test[:, 0], y=x_test[:, 1], c=y_pred, alpha=0.25)
```

```python
#Map with original data points

#create a plot with the data and predictions
fig, ax = plt.subplots()

ax.scatter(x_test[:, 0], y=x_test[:, 1], c=y_pred, alpha=0.25)

#Add the real data points to the scatter plot
colors = ['b', 'r', 'g']

y_col = data["Class ID"].apply(lambda x: colors[x])

ax.scatter(x=data["Weight"], y=data["Height"], c=y_col)
```
